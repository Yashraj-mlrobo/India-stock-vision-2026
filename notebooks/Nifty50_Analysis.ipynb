# Nifty 50 Analysis & Forecasting
This notebook contains the data ingestion, feature engineering, and visualization for the India Stock Vision 2026 project.

```
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../data/raw/NSEI.csv', header=[1,2], index_col=0, parse_dates=True)
df.columns = df.columns.get_level_values(0)
df['SMA_50'] = df['Close'].rolling(50).mean()
df['SMA_200'] = df['Close'].rolling(200).mean()
print('Data Loaded Successfully')
```

## Golden Cross Visualization
Identifying long-term trends in the NSE index.

```
plt.figure(figsize=(12,6))
plt.plot(df['Close'], alpha=0.5)
plt.plot(df['SMA_50'], label='50 SMA')
plt.plot(df['SMA_200'], label='200 SMA')
plt.title('Nifty 50 Trend Analysis')
plt.legend()
plt.show()
```
